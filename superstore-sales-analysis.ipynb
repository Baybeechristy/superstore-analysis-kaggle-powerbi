# Exploratory Data Analysis (EDA)

**Plot Sales Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'],bins=50, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

**Plot Profit Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Profit'],bins=50, kde=True)
plt.title('Distribution of Profit')
plt.xlabel('Profit Amount')
plt.ylabel('Frequency')
plt.show()
```

**Plot Discount Distribution**

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Discount'],bins=10, kde=False)
plt.title('Discount Distribution')
plt.xlabel('Discount')
plt.ylabel('count')
plt.show()
```

**Sales vs Profit**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category')
plt.title('Sales vs Profit by Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

Sales and profit are not strongly correlated. Some high-sale orders still result in losses, especially in the Furniture category. Technology generally performs better in both revenue and profit. Further analysis is needed to understand how discounting impacts profit

```python
#using heatmap to see which numeric columns are related and how strongly
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

The correlation matrix reveals a moderate positive relationship between Sales and Profit (0.48), and a negative relationship between Discount and Profit (-0.22), suggesting that higher discounting tends to reduce profitability. Interestingly, Quantity has little correlation with Profit, and Discounts donâ€™t significantly increase Sales, which raises concerns about the effectiveness of promotional strategies.

# Scatter Plot: Discount vs Profit

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category')
plt.title('Discount vs Profit by Category')
plt.xlabel('Discount (%)')
plt.ylabel('Profit (Â£)')
plt.axhline(0, color='gray', linestyle='--')  # Add zero-profit line
plt.show()
```

Discounts beyond 20% are generally unprofitable, especially in Furniture and Office Supplies. Technology handles discounting a bit better possibly due to higher margins.

# Box Plot: Profit Distribution by Discount


```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Discount', y='Profit')
plt.title('Profit Distribution Across Discount Levels')
plt.xlabel('Discount (%)')
plt.ylabel('Profit (Â£)')
plt.xticks(rotation=45)
plt.show()
```

The boxplot confirms that higher discounts are strongly associated with losses. The tipping point seems to be 20â€“30% discount, after which profitability collapses.

```python
df.groupby('Discount')['Profit'].mean().sort_index()
```

Visual analysis confirms a clear negative relationship between Discount and Profit. While modest discounts (0â€“20%) may still allow for positive returns, anything above 30% discount consistently leads to losses, especially in Furniture and Office Supplies. Technology shows slightly better resistance to discounting, but even there, profit drops off at high discount levels. This suggests that current promotional strategies may be too aggressive, and a data-driven discount policy should be implemented.

# Category & Sub-Category Performance

```python
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
category_summary
```

```python
#Sales by Category
plt.figure(figsize=(8,5))
sns.barplot(x=category_summary.index, y=category_summary['Sales'])
plt.title('Total Sales by Category')
plt.ylabel('Sales')
plt.xlabel('Category')
plt.tight_layout()
plt.show()
```

```python
#profit by category
plt.figure(figsize=(8,5))
sns.barplot(x=category_summary.index, y=category_summary['Profit'])
plt.title('Total Profit by Category')
plt.ylabel('Profit')
plt.xlabel('Category')
plt.tight_layout()
plt.show()
```

```python
subcategory_summary = df.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
subcategory_summary.head()
```

```python
#Sales by sub-category
plt.figure(figsize=(12,6))
sns.barplot(x=subcategory_summary.index, y=subcategory_summary['Sales'])
plt.title('Total Sales by Sub-Category')
plt.ylabel('Sales')
plt.xlabel('Sub-Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#profit by sub-category
plt.figure(figsize=(12,6))
sns.barplot(x=subcategory_summary.index, y=subcategory_summary['Profit'])
plt.title('Total Profit by Sub-Category')
plt.ylabel('Profit')
plt.xlabel('Sub-Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


### Category-Level Insights
- Technology leads in both total sales and profit, making it the most valuable product category overall.
- Office Supplies performs moderately in both areas.
- Furniture shows lower overall profit, despite having a decent level of sales. This may indicate high costs or excessive discounting in this category.

### Sub-Category-Level Insights
- Phones are the top-selling sub-category, consistent with their high unit prices and demand.
- Copiers are a high-profit sub-category, indicating strong margins and effective pricing.
- Tables show low or negative profit, despite reasonable sales â€” a key area to investigate for potential losses.
- Fasteners have low sales and low profit, indicating minimal business impact.

### Business Recommendations
- Focus more on Technology products, especially Phones and Copiers, to maximize both revenue and profit.
- Investigate the Furniture category, especially Tables, for unprofitable items or discount overuse.
- Consider reducing focus on Fasteners due to their low performance.
